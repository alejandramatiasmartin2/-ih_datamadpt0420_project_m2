*In order to make a good exploratory data analysis, I am going to follow several steps:*

##### Import all the necessary libraries: 


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import set_matplotlib_formats
from string import ascii_letters

set_matplotlib_formats('svg')
```

##### Import the **diamonds_train.csv** dataset and store it in a variable called diamonds:

```python
diamonds = pd.read_csv('./data/diamonds_train_1.csv')

diamonds
```

##### In order tu know the **Volume**, we need to apply the formula **Volume = length x width x height**: x (length), y (width), z (depth)

```python
diamonds['Volume'] = diamonds['x']* diamonds['y'] * diamonds['z']
diamonds

```

##### Now I am going to analyze the correlation of the Dataframe

```python
corr = diamonds.corr()
corr
```

### Visualizing summary statistics

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

##### Now I will graphically analyse the Number of records by category 

```python
fig=plt.figure(figsize=(8,8))
ax = diamonds['color'].value_counts().sort_values().plot(kind = 'bar')
```

```python
fig=plt.figure(figsize=(8,8))
ax = diamonds['cut'].value_counts().sort_values().plot(kind = 'bar')
```

```python
fig=plt.figure(figsize=(8,8))
ax = diamonds['clarity'].value_counts().sort_values().plot(kind = 'bar')
```

##### In terms of color, clarity and cut: G, Ideal and SI1 are the ones that have more number of records

##### It would be interesting to analyse the correlation between the different characteristics

```python
sns.set_theme(style="white")
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

##### Based on this graph, we can see that price and carat are highly correlated. Also the Volume


```python
fig=plt.figure(figsize=(32,16))
ax = sns.barplot(data=diamonds, x='carat', y ='price');
```

##### This graph shows that the higher the carat, the higher the price of the diamond tends to be

##### Now I will now focus on the colour feature

```python
fig=plt.figure(figsize=(8,8))
ax1 = sns.barplot(data=diamonds, x='color', y ='price')
```

```python
fig = plt.figure(figsize=(16, 8))
ax = sns.boxplot(x='color', y='price', data=diamonds.sort_values('color'));
```

##### The color with the highest average price is J, the least colorless of all the colors in our dataset

```python
fig = plt.figure(figsize=(8, 8))
ax = sns.boxplot(x='cut', y='price', data=diamonds.sort_values('cut'));
```

##### The cut with the highest average price is the Premium cut

```python
fig = plt.figure(figsize=(16, 8))
ax = sns.boxplot(x='clarity', y='price', data=diamonds.sort_values('clarity'));
```

##### The clarity with the highest average price is the SI2

```python
fig = plt.figure(figsize=(32, 8))
ax = sns.boxplot(x='Volume', y='price', data=diamonds.sort_values('Volume'));
```

##### In this graph we can see that the higher the volume, the higher the price tendency
